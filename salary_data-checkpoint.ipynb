**Relationship between years of experience increase and the Salary**

The salary roughly increases as years of experience increase.

```python
df.plot.scatter(x = 'Salary', y = 'Years of Experience')
```

```python
# As salary diff increases, years of experience increases
df.sort_values(by='Years of Experience', ascending=True).plot(kind="bar", figsize=(20, 6), x = "Salary", y = 'Years of Experience')
plt.title("Salary-Years of Experience", fontsize=14)
```

Salary increases as years of experience increases

```python
# Printing data distribution
df['Age'] = df['Age'].astype('int')
df['Years of Experience'] = df['Years of Experience'].astype('int')

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

df['Age'].plot(kind='hist', ax=axes[0, 0], title='Age')
df['Years of Experience'].plot(kind='hist', ax=axes[0, 1], title='YoE')
df['Salary'].plot(kind='hist', ax=axes[0, 2], title='Salary')
df['Gender'].value_counts().plot(kind='bar', ax=axes[1, 0], title='Gender')
df['Education Level'].value_counts().plot(kind='bar', ax=axes[1, 1], title='Education')
df['Job Title'].value_counts()[:20].plot(kind='bar', ax=axes[1, 2], title='Top 20 Job Titles')

fig.suptitle('Data Distribution', fontsize=16)
plt.tight_layout()
```

```python
# Pie chart: Proportion of Education Level and Age
plt.figure(figsize=(10 ,6))
df["Education Level"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Proportion of Education Level", fontsize=14)
plt.ylabel("")
plt.show()

plt.figure(figsize=(10 ,10))
df["Age"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Proportion of Age", fontsize=10)
plt.ylabel("")
plt.show()
```

**Gender Analysis**

The male salary ratio is little bit higher than female salary ratio. 
Female earning most under customer service manager job title while male earning more under CEO and chief technology officer. 
Female highest salary is 130000 and male highest salary is 250000. 
In business analyst job title where, female salary is higher than male salary which is 80000 and 75000 respectively. 
But overall male salary is higher than female salary.

```python
# Printing the analysis of salaries by gender
sns.boxplot(x='Gender', y='Salary', data=df)
plt.title('Salary by Gender')
plt.show()
```

```python
# Categorical scatterplot for each job title
g = sns.catplot(x='Gender', y='Salary', col='Job Title', 
                data=df, kind='box', col_wrap=3, height=4, aspect=1.2)

# title name for each subplot
g.set_titles('{col_name}')

# Set the y-axis label
g.set_ylabels('Salary')

# Set the x-axis label
g.set_xlabels('Gender')

# Set the overall title
plt.suptitle('Salary by Gender and Job Title', y=1.02)

plt.show()
```

```python
# Printing mean for female salaries for 10 entries
df3 = df[df['Gender'] == 'Female']
female_jobs = df3.groupby('Job Title')['Salary'].mean().head(10)
female_jobs
```

```python
# female_groupby.plot(kind='scatterplot')
plt.figure(figsize=(8,5))
ax = sns.scatterplot(data=female_jobs)
plt.xticks(rotation=65)
plt.rcParams['lines.markersize'] = 30           
plt.title('Female Top Jobs Salaries')
```

```python
# Printing mean for male salaries for 10 entries
df3 = df[df['Gender'] == 'Male']
male_jobs = df3.groupby('Job Title')['Salary'].mean().head(10)
male_jobs
```

```python
# male_groupby.plot(kind='scatterplot')
plt.figure(figsize=(8,5))
ax = sns.scatterplot(data=male_jobs)
plt.xticks(rotation=65)
plt.rcParams['lines.markersize'] = 30           
plt.title('Male Top Jobs Salaries')
```
